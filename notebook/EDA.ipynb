### Some categorical data to check at first glance

```python
interesting_categorical_var = [
    'CODE_GENDER',
    'NAME_EDUCATION_TYPE',
    'OCCUPATION_TYPE',
    'WEEKDAY_APPR_PROCESS_START',
    'NAME_INCOME_TYPE',
    'NAME_FAMILY_STATUS',
]
```

```python
for column in interesting_categorical_var:
    print(f'Checking {column}...')
    result_df = training_base[[column, 'TARGET']].groupby(column).agg({
        len,
        sum,
    })
    result_df.columns=['Total', 'Defect']
    result_df['Rate'] = result_df.apply(
        lambda _: _['Defect'] / _['Total'] if _['Total'] > 0 else np.nan,
        axis=1
    )
    print(result_df.sort_values(by=['Rate', 'Total'], ascending=[False, False]))
    print('-'*50)
```

Some findings here:

- Males have less application but higher defect rate
- Person with less education have higher defect rate
- Widow, surprisingly have the lower defect rate
- Person applied the loan during weekdays seem to have little bit higher defect rate
- Person with lower skill level at work have lower defect rate
